```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Read the csv file into a pandas DataFrame
pop = pd.read_csv('../Resources/population_2011_2018.csv')
pop.head()
```

```python
df = pd.DataFrame(columns=['FIPS','State','Area_Name','Year','Population','Births','Deaths',
                            'InternationalMig','DomesticMig','CivilianLaborForce','UnemploymentRate'])
pop.State.unique().tolist()
```

```python
State = 'WY'
data= pop.groupby('State').get_group(State).reset_index() 
    
X1 = data.Year.values.reshape(-1, 1)
y1 = data.Population.values.reshape(-1, 1)
y2 = data.Births.values.reshape(-1, 1)
y3 = data.Deaths.values.reshape(-1, 1)
y4 = data.InternationalMig.values.reshape(-1, 1)
y5 = data.DomesticMig.values.reshape(-1, 1)
y6 = data.CivilianLaborForce.values.reshape(-1, 1)
y7 = data.UnemploymentRate.values.reshape(-1, 1)
X=X1
Year2019=2019
Year2020=2020
Year2021=2021
FIPS=data.FIPS[1]
Area_Name=data.Area_Name[1]
model = LinearRegression()

model.fit(X, y2)
Births2019 = model.predict([[Year2019]])[(0)]
Births2020 = model.predict([[Year2020]])[(0)]
Births2021 = model.predict([[Year2021]])[(0)]    
model.fit(X, y3)
Deaths2019 = model.predict([[Year2019]])[(0)]
Deaths2020 = model.predict([[Year2020]])[(0)]
Deaths2021 = model.predict([[Year2021]])[(0)]
model.fit(X, y4)
InternationalMig2019 = model.predict([[Year2019]])[(0)]
InternationalMig2020 = model.predict([[Year2020]])[(0)]
InternationalMig2021 = model.predict([[Year2021]])[(0)]
model.fit(X, y5)
DomesticMig2019 = model.predict([[Year2019]])[(0)]
DomesticMig2020 = model.predict([[Year2020]])[(0)]
DomesticMig2021 = model.predict([[Year2021]])[(0)]
model.fit(X, y6)
CivilianLaborForce2019 = model.predict([[Year2019]])[(0)]
CivilianLaborForce2020 = model.predict([[Year2020]])[(0)]
CivilianLaborForce2021 = model.predict([[Year2021]])[(0)]
model.fit(X, y7)
UnemploymentRate2019 = model.predict([[Year2019]])[(0)]
UnemploymentRate2020 = model.predict([[Year2020]])[(0)]
UnemploymentRate2021 = model.predict([[Year2021]])[(0)]

X2019=pd.DataFrame({"Births": Births2019,"Deaths": Deaths2019,"InternationalMig": 
                     InternationalMig2019,"DomesticMig": DomesticMig2019,
                    "CivilianLaborForce": CivilianLaborForce2019,
                     "UnemploymentRate": UnemploymentRate2019})
X2020=pd.DataFrame({"Births": Births2020,"Deaths": Deaths2020,"InternationalMig": 
                     InternationalMig2020,"DomesticMig": DomesticMig2020,
                    "CivilianLaborForce": CivilianLaborForce2020,
                     "UnemploymentRate": UnemploymentRate2020})
X2021=pd.DataFrame({"Births": Births2021,"Deaths": Deaths2021,"InternationalMig": 
                     InternationalMig2021,"DomesticMig": DomesticMig2021,
                    "CivilianLaborForce": CivilianLaborForce2021,
                     "UnemploymentRate": UnemploymentRate2021})

X = pop[["Births","Deaths","InternationalMig","DomesticMig","CivilianLaborForce","UnemploymentRate"]]
y = pop["Population"].values.reshape(-1, 1)
model.fit(X, y)
predictions2019 = int(model.predict(X2019)[(0,0)])
predictions2020 = int(model.predict(X2020)[(0,0)])
predictions2021 = int(model.predict(X2021)[(0,0)])

df = df.append({'FIPS':FIPS,'State':State,'Area_Name':Area_Name,'Year':Year2019,
                'Population':predictions2019,'Births':Births2019[0],'Deaths':Deaths2019[0],
                'InternationalMig':InternationalMig2019[0],'DomesticMig':DomesticMig2019[0],
                'CivilianLaborForce':CivilianLaborForce2019[0],
                'UnemploymentRate':UnemploymentRate2019[0]}, ignore_index=True)
df = df.append({'FIPS':FIPS,'State':State,'Area_Name':Area_Name,'Year':Year2020,
                'Population':predictions2020,'Births':Births2020[0],'Deaths':Deaths2020[0],
                'InternationalMig':InternationalMig2020[0],'DomesticMig':DomesticMig2020[0],
                'CivilianLaborForce':CivilianLaborForce2020[0],
                'UnemploymentRate':UnemploymentRate2020[0]}, ignore_index=True)
df = df.append({'FIPS':FIPS,'State':State,'Area_Name':Area_Name,'Year':Year2021,
                'Population':predictions2021,'Births':Births2021[0],'Deaths':Deaths2021[0],
                'InternationalMig':InternationalMig2021[0],'DomesticMig':DomesticMig2021[0],
                'CivilianLaborForce':CivilianLaborForce2021[0],
                'UnemploymentRate':UnemploymentRate2021[0]}, ignore_index=True)
df.to_csv('../Resources/population3.csv')
df.head()
```
