```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame
# 1
# lsd = pd.read_csv('../Resources/lsd.csv')
# lsd.head()
# 2
# brains = pd.read_csv('../Resources/brain.csv')
# brains.head()
# 3
# foam = pd.read_csv('../Resources/foam.csv')
# foam.head()
# 4
smoking = pd.read_csv('../Resources/smoking.csv')
smoking.head()
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this
# 1
# X = lsd.tc.values.reshape(-1, 1)
# y = lsd.score.values.reshape(-1, 1)
# print("Shape: ", X.shape, y.shape)
# X
# 2
# X = brains["weight"].values.reshape(-1, 1)
# y = brains["size"].values.reshape(-1, 1)
# print("Shape: ", X.shape, y.shape)
# 3
# X = foam[["foam", "beer"]]
# y = foam["time"].values.reshape(-1, 1)
# print(X.shape, y.shape)
# 4
# Use Pandas get_dummies to convert categorical data
smoking = pd.get_dummies(smoking)
smoking.head()
```

```python
# Assign the data to X and y
X = smoking[['Age', 'Height', 'Sex_Female', 'Sex_Male', 'Smoker_Current', 'Smoker_Non']]
y = smoking["FEV"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
print(X,y)
```

```python
# Plot the data
# 1
# plt.scatter(X, y)
# 2
# plt.scatter(X, y)
# plt.xlabel("Brain Weight")
# plt.ylabel("Head Size")
```

```python
# Use sklearn's `train_test_split` to split the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train, X_test, y_train, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
print(X_train_scaled,X_test_scaled,y_train_scaled,y_test_scaled)
```

```python
# Create the model and fit the model to the data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the data
# 1
# model.fit(X, y)
# 2
model.fit(X_train, y_train)
```

```python
# Print the coefficient and the intercept for the model
# 1
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
# Note: we have to transform our min and max values 
# 1
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max
# 1
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
# Plot X and y using plt.scatter
# Plot the model fit line using [x_min[0], x_max[0]], [y_min[0], y_max[0]]
# 1
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```

```python
# Calculate the mean_squared_error and the r-squared value
from sklearn.metrics import mean_squared_error, r2_score
predicted = model.predict(X_test)
# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
# Call the `score` method on the model to show the r2 score
model.score(X_test, y_test)
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data
# 3
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Plot the Residuals for the Training and Testing data
# 3
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```
